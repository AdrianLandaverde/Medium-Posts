# 01. Libraries and Data

```python
import pandas as pd
import numpy as np
import statsmodels.api as sm
from scipy import stats
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', 10)
```

```python
df_co2= pd.read_csv('../../data/co2_emissions.csv')
df_co2
```

```python
df_electric= pd.read_csv('../../data/electric_consumption.csv')
df_electric
```

```python
df_forest= pd.read_csv('../../data/forest_area.csv')
df_forest
```

```python
df_population= pd.read_csv('../../data/population.csv')
df_population
```

# 02. Data Cleaning

```python
df_co2= df_co2[df_co2['Country Name']=='Mexico']
df_co2= df_co2.iloc[:, 4:].T.dropna().reset_index()
df_co2.columns= ['Year', 'CO2 Emissions']
df_co2
```

```python
df_electric= df_electric[df_electric['Country Name']=='Mexico']
df_electric= df_electric.iloc[:, 4:].T.dropna().reset_index()
df_electric.columns= ['Year', 'Electric Consumption']
df_electric
```

```python
df_forest= df_forest[df_forest['Country Name']=='Mexico']
df_forest= df_forest.iloc[:, 4:].T.dropna().reset_index()
df_forest.columns= ['Year', 'Forest Area']
df_forest
```

```python
df_population= df_population[df_population['Country Name']=='Mexico']
df_population= df_population.iloc[:, 4:].T.dropna().reset_index()
df_population.columns= ['Year', 'Population']
df_population
```

```python
df_mexico= pd.merge(df_co2, df_electric, on='Year')
df_mexico= pd.merge(df_mexico, df_forest, on='Year')
df_mexico= pd.merge(df_mexico, df_population, on='Year')
df_mexico= df_mexico.set_index('Year')
df_mexico.columns= ['co2_emissions', 'electric_consumption', 'forest_area', 'population']
df_mexico
```

# 03. Model

```python
X= df_mexico[['electric_consumption', 'forest_area', 'population']]
Y= df_mexico['co2_emissions']
X= sm.add_constant(X)
model= sm.OLS(Y, X).fit()
```

```python
model.summary()
```

```python
y_pred= model.predict(X)
```

# 04. Calculate the summary statistics

### 04.1 Summary statistics

```python
n_obsertions= df_mexico.shape[0]
k= X.shape[1]-1
df_residuals= n_obsertions - k -1
df_models= k
```

### 04.2 R Squared

```python
residuals= Y - y_pred
ssr= np.dot(residuals, residuals)
print('Sum of Squared Residuals:', ssr)
```

```python
centered_y= Y - Y.mean()
sst= np.dot(centered_y, centered_y)
print('Total Sum of Squares:', sst)
```

```python
r2= 1 - ssr/sst
print('R^2:', r2)
```

### 04.3 Adjusted R Squared

```python
adj_r2= 1 - (1-r2)*(n_obsertions-1)/(n_obsertions-k-1)
print('Adjusted R^2:', adj_r2)
```

### 04.4 F-Statistic

```python
explained_sum_squares= sst - ssr
```

```python
mse_model= explained_sum_squares/k
print('Mean Squared Error of the Model:', mse_model)
```

```python
mse_residuals= ssr/df_residuals
print('Mean Squared Error of the Residuals:', mse_residuals)
```

```python
fvalue= mse_model/mse_residuals
print('F-Value:', fvalue)
```

```python
p_value= stats.f.sf(fvalue, k, df_residuals)
print('P-Value:', p_value)
```

### 04.5 Log Likelihood

```python
log_likelihood= -n_obsertions/2 * (1 + np.log(2*np.pi) + np.log(ssr/n_obsertions))
print('Log Likelihood:', log_likelihood)
```

### 04.6 Aikaikes Information Criterion (AIC)

```python
aic_model= -2*log_likelihood + 2*(k+1)
print('AIC:', aic_model)
```

### 04.7 Bayesian Information Criterion (BIC)

```python
bic_model= -2*log_likelihood + (k+1)*np.log(n_obsertions)
print('BIC:', bic_model)
```

### 04.8 Coefficients

```python
X_squrared= np.dot(X.T, X)
X_squrared_inv= np.linalg.inv(X_squrared)
mult_x= np.dot(X_squrared_inv, X.T)
coefficients= np.dot(mult_x, Y)
print('Coefficients:', coefficients)
```

### 04.9 Coefficient Standard Errors

```python
resid_variance= ssr/df_residuals
print('Residual Variance:', resid_variance)
```

```python
cov_matrix= resid_variance * X_squrared_inv
print('Covariance Matrix:', cov_matrix)
```

```python
variances= np.diagonal(cov_matrix)
std_errors= np.sqrt(variances)
print('Standard Errors:', std_errors)
```

```python
t_stats= coefficients/std_errors
print('T-Statistics:', t_stats)
```

```python
p_values= stats.t.sf(np.abs(t_stats), df_residuals)*2
print('P-Values:', p_values)
```

### 04.10 Confidence Intervals

```python
upper_interval= coefficients + stats.t.ppf(0.975, df_residuals)*std_errors
lower_interval= coefficients - stats.t.ppf(0.975, df_residuals)*std_errors
print('Confidence Interval:', lower_interval, upper_interval)
```

### 04.11 Omni Test

```python
omni_test= stats.normaltest(residuals, axis=0)
print('Omnibus Test:', omni_test)
```

### 04.12 Skewness & Kurtosis

```python
skewness= stats.skew(residuals, axis=0)
print('Skewness:', skewness)
```
